```python
import sympy as sp
w = sp.symbols('w', cls=sp.Function)
C1, C2, C3, C4 = sp.symbols('C1 C2 C3 C4')
x = sp.symbols('x')
L = 10
EI = sp.nsimplify(4+x) #TO DO: investigate which functions work and which don't
q = sp.nsimplify((x-4)**2-2.5*x) #TO DO: investigate which functions work and which don't

sp.plotting.plot(EI,(x,0.0001,L))
sp.plotting.plot(q,(x,0,L))

diffeq = sp.Eq(EI*sp.diff(w(x),x,4),q)
display(diffeq)

w = sp.dsolve(diffeq)
w = w.rhs
display(w)

phi = -sp.diff(w, x)
kappa = sp.diff(phi, x)
M = EI * kappa
V = sp.diff(M, x)

eq1  = sp.Eq(w.subs(x , 0) , 0)
eq2  = sp.Eq(M.subs(x , 0) , 0)
eq3  = sp.Eq(w.subs(x , L) , 0)
eq4  = sp.Eq(phi.subs(x , L) , 0)

sol = sp.solve((eq1,eq2,eq3,eq4) ,
               (C1 ,C2 ,C3 ,C4))
w_sol = w.subs(sol)
M_sol = M.subs(sol)

sp.plot(w_sol,(x,0,L))
sp.plot(M_sol,(x,0,L));
```
