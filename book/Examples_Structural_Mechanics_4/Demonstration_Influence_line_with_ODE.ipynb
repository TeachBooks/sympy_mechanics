# Influence line displacement at D

```python
import sympy as sp
```

```python
x = sp.symbols('x')
w_func = sp.symbols('w_func',cls=sp.Function)      
EI = 5*1000
```

```python
C1, C2, C3, C4, C5, C6, C7, C8 = sp.symbols('C1, C2, C3, C4, C5, C6, C7, C8')
```

```python
q_1 = sp.SingularityFunction(x,3,-1)
q_2 = 0
```

```python
V_1 = -sp.integrate(q_1,x)+C1
M_1 = sp.integrate(V_1,x)+C2
kappa_1 = M_1 / EI
phi_1 = sp.integrate(kappa_1,x)+C3
w_1 = -sp.integrate(phi_1,x)+C4

V_2 = -sp.integrate(q_2,x)+C5
M_2 = sp.integrate(V_2,x)+C6
kappa_2 = M_2 / EI
phi_2 = sp.integrate(kappa_2,x)+C7
w_2 = -sp.integrate(phi_2,x)+C8
```

```python
eq1 = sp.Eq(w_1.subs(x,0),0)
eq2 = sp.Eq(M_1.subs(x,0),0)
eq3 = sp.Eq(w_1.subs(x,5),0)
eq4 = sp.Eq(M_1.subs(x,5),M_2.subs(x,5))
eq5 = sp.Eq(phi_1.subs(x,5),phi_2.subs(x,5))
eq6 = sp.Eq(w_2.subs(x,5),0)
eq7 = sp.Eq(w_2.subs(x,10),0)
eq8 = sp.Eq(M_2.subs(x,10),0)
sol = sp.solve([eq1,eq2,eq3,eq4,eq5,eq6,eq7,eq8],[C1,C2,C3,C4,C5,C6,C7,C8])
display(sol)
```

```python
w_sol_1 = w_1.subs(sol)
w_sol_2 = w_2.subs(sol)
```

```python
sp.plot((-w_sol_1,(x,0,5)),((-w_sol_2,(x,5,10))))
```

# Influence line support reaction at A

```python
q_1 = 0
V_1 = -sp.integrate(q_1,x)+C1
M_1 = sp.integrate(V_1,x)+C2
kappa_1 = M_1 / EI
phi_1 = sp.integrate(kappa_1,x)+C3
w_1 = -sp.integrate(phi_1,x)+C4
```

```python
eq1 = sp.Eq(w_1.subs(x,0),1)
eq2 = sp.Eq(M_1.subs(x,0),0)
eq3 = sp.Eq(w_1.subs(x,5),0)
eq4 = sp.Eq(M_1.subs(x,5),M_2.subs(x,5))
eq5 = sp.Eq(phi_1.subs(x,5),phi_2.subs(x,5))
eq6 = sp.Eq(w_2.subs(x,5),0)
eq7 = sp.Eq(w_2.subs(x,10),0)
eq8 = sp.Eq(M_2.subs(x,10),0)
sol = sp.solve([eq1,eq2,eq3,eq4,eq5,eq6,eq7,eq8],[C1,C2,C3,C4,C5,C6,C7,C8])
display(sol)
```

```python
w_sol_1 = w_1.subs(sol)
w_sol_2 = w_2.subs(sol)
```

```python
sp.plot((-w_sol_1,(x,0,5)),((-w_sol_2,(x,5,10))))
```
